Exploratory Data Analysis on the Kaggle Credit Card Fraud dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 6)

# Load dataset
DATA_PATH = os.path.join(os.path.dirname(__file__), '..', 'data', 'raw', 'creditcard.csv')
df = pd.read_csv(DATA_PATH)
print(f"Shape: {df.shape}")
df.head()
```

## 1. Basic Info

```python
print("=== Dataset Info ===")
print(df.info())
print("\n=== Missing Values ===")
print(df.isnull().sum().sum())
print("\n=== Basic Stats ===")
df.describe()
```

## 2. Class Imbalance

```python
fraud_count = df['Class'].value_counts()
print(f"Legitimate (0): {fraud_count[0]} ({fraud_count[0]/len(df)*100:.2f}%)")
print(f"Fraud (1):     {fraud_count[1]} ({fraud_count[1]/len(df)*100:.4f}%)")

fig, ax = plt.subplots()
fraud_count.plot(kind='bar', color=['steelblue', 'crimson'], ax=ax)
ax.set_title('Class Distribution')
ax.set_xlabel('Class (0=Legit, 1=Fraud)')
ax.set_ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('../docs/class_distribution.png', dpi=150)
plt.show()
```

## 3. Transaction Amount Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# All transactions
df[df['Class']==0]['Amount'].hist(bins=50, ax=axes[0], alpha=0.7, color='steelblue')
df[df['Class']==1]['Amount'].hist(bins=50, ax=axes[1], alpha=0.7, color='crimson')

axes[0].set_title('Legitimate Transactions - Amount')
axes[1].set_title('Fraud Transactions - Amount')
plt.tight_layout()
plt.savefig('../docs/amount_distribution.png', dpi=150)
plt.show()

print("Legit amount stats:", df[df['Class']==0]['Amount'].describe())
print("\nFraud amount stats:", df[df['Class']==1]['Amount'].describe())
```

## 4. Time Analysis

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
df[df['Class']==0]['Time'].hist(bins=50, ax=axes[0], alpha=0.7, color='steelblue')
df[df['Class']==1]['Time'].hist(bins=50, ax=axes[1], alpha=0.7, color='crimson')
axes[0].set_title('Legitimate - Time')
axes[1].set_title('Fraud - Time')
plt.tight_layout()
plt.savefig('../docs/time_distribution.png', dpi=150)
plt.show()
```

## 5. Feature Correlations with Class

```python
# Top correlations with Class
corrs = df.corr()['Class'].drop('Class').sort_values(key=abs, ascending=False)
print("Top correlations with Class:")
print(corrs.head(10))

fig, ax = plt.subplots(figsize=(10, 6))
corrs.plot(kind='barh', ax=ax, color='steelblue')
ax.set_title('Feature Correlations with Class')
plt.tight_layout()
plt.savefig('../docs/feature_correlations.png', dpi=150)
plt.show()
```

## 6. Key Insights Summary

- **Extreme imbalance**: ~0.17% fraud transactions
- **Amount**: Fraud transactions tend to have higher amounts on average
- **V1-V28**: PCA-transformed features (confidential)
- **Time**: Seconds elapsed from first transaction
- **Strategy**: Use SMOTE/RandomUnderSampler + threshold tuning
